```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# machine learning
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler,
    MinMaxScaler, 
    PolynomialFeatures,
    KBinsDiscretizer
)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

```python

fin = pd.read_csv('data/Train.csv')
fin.head()
```

```python
fin = fin.drop('relationship_with_head', axis=1)
fin.head()
```

```python
sns.countplot(x='bank_account', hue= 'education_level', data=fin)
```

```python
le = LabelEncoder()
fin['bank_account'] = le.fit_transform(fin['bank_account'])



X = fin.drop('bank_account', axis=1)
y = fin['bank_account']
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y , test_size= 0.25, stratify=y)
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False)
ohe.set_output(transform ='pandas')
trans = ohe.fit_transform(X_train[['location_type', 'cellphone_access', 'gender_of_respondent',
                'marital_status','education_level',
                 'job_type']])
trans1 = ohe.transform(X_test[['location_type', 'cellphone_access', 'gender_of_respondent',
                              'marital_status','education_level', 
                              'job_type']])
trans1
```

```python
df_train = pd.DataFrame(trans)

df_test = pd.DataFrame(trans1)
```

```python
X_train = X_train.join(df_train)
X_test = X_test.join(df_test)
```

```python
y_test.value_counts()
```

```python
X_train = X_train.drop(['location_type', 'cellphone_access', 'gender_of_respondent',
                             'marital_status','education_level', 
                              'job_type','country', 'year', 'uniqueid'], axis=1)
X_test = X_test.drop(['location_type', 'cellphone_access', 'gender_of_respondent',
                              'marital_status','education_level', 
                              'job_type','country', 'year', 'uniqueid'], axis=1)
```

```python

log = LogisticRegression(class_weight= 'balanced')

log.fit(X_train, y_train)
```

```python
y_pred = log.predict(X_test)
y_pred_tr = log.predict(X_train)
```

```python
print(classification_report(y_pred, y_test))
```

```python
Knn = KNeighborsClassifier(n_neighbors =5 ,metric='manhattan')
Knn.fit(X_train, np.ravel(y_train))
```

```python
y_knn_test = Knn.predict(X_test)
y_knn_train = Knn.predict(X_train)
```

```python
print(classification_report(y_knn_test, y_test))
```

```python
print(classification_report(y_knn_train, y_train))
```

```python
print(confusion_matrix(y_knn_test, y_test))
```
